```python
# df = pd.read_excel('NSSE24 Data (UNT Challenge) Akula.xlsx')
import pandas as pd

df = pd.read_excel("Book2.xlsx", engine="openpyxl")
print(df.head())
```

```python
# Examine data types and identify numerical and categorical columns
print("Data Types:\n", df.dtypes)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("\nMissing Values Percentage:\n", missing_percentage)
```

```python
# columns with more than 1 missing value and their missing value percentages
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_columns = missing_percentage[missing_percentage > 1].index
print("\nColumns with More than 1 Missing Value and Their Missing Value Percentages:\n", missing_columns)
```

```python
df.head()
```

```python
# Apply stripping and replace empty strings only on object columns
df = df.apply(lambda col: col.astype(str).str.strip().replace('', pd.NA) if col.dtype == 'object' else col)

# Convert numeric-like object columns back to proper numeric type
df = df.apply(lambda col: pd.to_numeric(col, errors='ignore') if col.dtype == 'object' else col)

print(df.dtypes)  # Verify column data types
print(df.head())   # Check if data is correctly retained
```

```python
df.head()
```

```python
# Function to check if a column contains any string values (excluding NaN)
def check_string_columns(df):
    string_cols = []
    
    for col in df.columns:
        if df[col].dropna().apply(lambda x: isinstance(x, str)).any():
            string_cols.append(col)
    
    return string_cols

# Get list of columns with string values
string_columns = check_string_columns(df)

print("Columns that contain string values:", string_columns)
```

```python
# impute MAJfirst', 'MAJsecond', 'gi_another_txt', 'so_another_txt', 'gr_another_txt', 'group1', 'group4', gi_another_txt, so_another_txt with mode
df['MAJfirst'] = df['MAJfirst'].fillna(df['MAJfirst'].mode()[0])
df['MAJsecond'] = df['MAJsecond'].fillna(df['MAJsecond'].mode()[0])
df['gr_another_txt'] = df['gr_another_txt'].fillna(df['gr_another_txt'].mode()[0])
df['group1'] = df['group1'].fillna(df['group1'].mode()[0])
df['group4'] = df['group4'].fillna(df['group4'].mode()[0])
df['gi_another_txt'] = df['gi_another_txt'].fillna(df['gi_another_txt'].mode()[0])
df['so_another_txt'] = df['so_another_txt'].fillna(df['so_another_txt'].mode()[0])
```

```python
# impute NaN with median of all columns except MAJfirst', 'MAJsecond', 'gi_another_txt', 'so_another_txt', 'gr_another_txt', 'group1', 'group4'
df = df.apply(lambda col: col.fillna(col.median()) if col.name not in ['MAJfirst', 'MAJsecond', 'gi_another_txt', 'so_another_txt', 'gr_another_txt', 'group1', 'group4'] else col)
```

```python
# print dtypes of all columns
print(df.dtypes)
```

```python
#print columns with object type
print(df.select_dtypes(include=['object']).columns)
```

```python
#number of columns in df that are objects type 
print(len(df.select_dtypes(include=['int']).columns))
```
